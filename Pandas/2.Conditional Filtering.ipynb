# Conditional Filtering

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('tips.csv')
```

```python
df.head(3)
```

## Conditions

```python
# Some condition

# df['price_per_person'] > 16

#cot = df['price_per_person'] > 16

#df[cot]
```

```python
df[df['price_per_person'] > 16]
```

```python
df[df['sex'] == 'Female']
```

## More option in conditions

* OR | (either condition is True)
* AND & (both conditions must be True)
* You can also use the ~ operator as a NOT operation

```python
df[(df['price_per_person'] > 12) & (df['sex']=='Male')]
```

```python
df[(df['price_per_person'] > 12) & ~(df['sex']=='Male')]
```

## Conditional Operator isin()

We can use .isin() operator to filter by a list of options.

```python
options = ['Sat','Sun']
df['day'].isin(options)
```

```python
df[df['day'].isin(['Thur' ,'Fri '])]
```
